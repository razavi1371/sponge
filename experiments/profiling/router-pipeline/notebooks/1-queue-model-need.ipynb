```python
experiment_ids = key_config_df[
    key_config_df['load'] == 10][
    'experiment_id'].tolist()

metadata_columns = [
    'experiment_id',
    'task_0_cpu_request',
    'task_0_model_variant',
    'task_0_replica',
    'task_0_max_batch_size',
    'task_1_cpu_request',
    'task_1_model_variant',
    'task_1_replica',
    'task_1_max_batch_size',
    ]

selected_results_columns = [
    'client_to_pipeline_latencies_avg',
    'task_0_model_latencies_avg',
    'task_0_to_task_1_latencies_avg',
    'task_1_model_latencies_avg',
    ]


output = loader.table_maker(
    experiment_ids=experiment_ids,
    metadata_columns=metadata_columns,
    results_columns=selected_results_columns)
pd.set_option('display.max_rows', None)
output['model_latencies_sum'] = output['task_0_model_latencies_avg'] + output['task_1_model_latencies_avg']
output['e2e_latencies'] = output['task_0_model_latencies_avg'] + output['task_1_model_latencies_avg'] + output['task_0_to_task_1_latencies_avg']
```
