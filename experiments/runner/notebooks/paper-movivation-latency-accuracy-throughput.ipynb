```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np

# get an absolute path to the directory that contains parent files
__file__ = globals()["_dh"][0]
project_dir = __file__ = globals()["_dh"][0]
sys.path.append(os.path.normpath(os.path.join(project_dir, "..", "..", "..")))

from experiments.utils.constants import FINAL_RESULTS_PATH, FIGURES_PATH
```

```python
font = {"size": 12}
plt.rc("font", size=12)
plt.rc("axes", titlesize=12)
plt.rcParams["pdf.fonttype"] = 42
plt.rcParams["ps.fonttype"] = 42
fig, axs = plt.subplots(2, 1, figsize=(6, 4))
ax1 = axs[0]
ax2 = axs[1]
ax1_acc = ax1.twinx()
ax2_acc = ax2.twinx()
ax1.grid(axis="y", color="gray", linestyle="dashed")
ax2.grid(axis="y", color="gray", linestyle="dashed")
ax2.set_xlabel("ResNet variant")
ax1.set_ylabel("Throughput (RPS)")
ax1_acc.set_ylabel("Accuracy (%)")

ax2.set_ylabel("Latency (ms)")
ax2_acc.set_ylabel("Accuracy (%)")

ax1.bar(["18", "34", "50", "101", "152"], [19, 11, 9, 5, 4], color="#2c7fb8")
ax1.set_xticklabels([])
ax1_acc.plot(
    np.arange(5),
    [69.75, 73.31, 76.13, 77.37, 78.31],
    color="black",
    marker="D",
    label="Accuracy",
)

ax2.bar(["18", "34", "50", "101", "152"], [70, 110, 135, 240, 300], color="#41b6c4")
ax2_acc.plot(
    np.arange(5),
    [69.75, 73.31, 76.13, 77.37, 78.31],
    color="black",
    marker="D",
    label="Accuracy",
)

plt.legend(loc="right")
plt.savefig(
    dpi=600,
    format="pdf",
    bbox_inches="tight",
    fname=f"{FIGURES_PATH}/two_metrics.pdf",
)
```
