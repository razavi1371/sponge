```python
import pandas as pd
import os
import sys
from pprint import PrettyPrinter
import matplotlib.pyplot as plt
from typing import Dict, List

pp = PrettyPrinter(indent=4)
from barazmoon.twitter import twitter_workload_generator

# get an absolute path to the directory that contains parent files
__file__ = globals()["_dh"][0]
project_dir = __file__ = globals()["_dh"][0]
sys.path.append(os.path.normpath(os.path.join(project_dir, "..", "..", "..")))

from experiments.utils.constants import FINAL_RESULTS_PATH
from experiments.utils.parser import AdaptationParser
```

```python
metaseries = 0
start = 120
end = 200
serieses = list(range(start, end + 1))

series_paths = {}
loaders = {}

for series in serieses:
    print(f"series: {series}")
    try:
        series_path = os.path.join(
            FINAL_RESULTS_PATH, "metaseries", str(metaseries), "series", str(series)
        )
        series_paths[series] = series_path
        loaders[series] = AdaptationParser(
            series_path=series_path, model_name="video", type_of="router_pipeline"
        )
    except FileNotFoundError:
        pass
```

```python
for series, loader in loaders.items():
    configs_exp = loader.load_configs()
    print(f"series {series} metadata:")
    try:
        config = configs_exp["0.yaml"]
    except KeyError:
        continue
    # pp.pprint(config)
    print(f"{config['metadata']}")
    try:
        print(f"outcome: {config['outcome']}")
    except KeyError:
        pass
    try:
        print(f"todo: {config['todo']}")
    except KeyError:
        pass
    print("\n")
```
