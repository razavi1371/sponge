```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x-axis range and exponential growth rate
x = np.linspace(0, 5, 100)
y = np.exp(x)

# Plot the exponential growth
plt.plot(y, x)

# Label the x-axis and y-axis
plt.xlabel('Throughput')
plt.ylabel('Latency')

# Show the plot
plt.show()
```
